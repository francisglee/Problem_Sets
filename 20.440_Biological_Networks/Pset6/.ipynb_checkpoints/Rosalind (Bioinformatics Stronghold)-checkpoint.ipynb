```python
# Rosalind DNA: Counting DNA Molecules

def countnucleotides(sequence):
    """
    Input: DNA sequence, uppercase string
    Output: 4 integers: number of 'A', 'C', 'G', 'T'
    """
    A = 0 
    C = 0
    G = 0
    T = 0
    
    for ch in sequence:
        if ch == 'A':
            A += 1
        elif ch == 'C':
            C += 1
        elif ch == 'G':
            G += 1
        elif ch == 'T':
            T += 1
    
    return A, C, G, T

def countnucleotides_complete():
    """
    Input: None; user is asked for sequence.
    Output: Frequency of nucleotides in sequence.  ie: A C G T: 23 54 25 24
    contraints: strings must be in upper case
    """
    sequence = getinput()
    if sequence == "ERROR":
        print "No sequence to analyze.  Please retry."
    else:
        newsequence = sequence.upper()
        print "A C G T: " + str(countnucleotides(newsequence))
    print 
        
countnucleotides_complete()

# Two line solution:
# nucleotides = file('rosalind_dna.txt').read()
# print [nucleotides.count(c) for c in 'ACGT']
```

```python
# Rosalind RNA: Transcribing DNA into RNA

def transcription(template):
    """
    input: DNA sequence, template, in string format
    output: RNA sequence, transcript, in string format
    contraints: strings must be in upper case
    """
    transcript = ""
    for c in template:
        if c != 'T':
            transcript += c
        elif c == 'T':
            transcript += 'U'
    
    return transcript

def transcription_complete():
    """
    input: None; user is prompted for file or pasted string/values that contains a DNA sequence
    output: RNA sequence, transcript, in string format, all Upper Case
    """
    sequence = getinput()
    if sequence == "ERROR":
        print "No sequence to analyze.  Please retry."
    else:
        newsequence = sequence.upper()
        print transcription(newsequence)

transcription_complete()

# One line solution:
# print raw_input().replace('T','U')
            
```

```python
# Rosalind REVC: Complementing a Strand of DNA

def dnacomplement(template):
    """
    input: A DNA string, template
    output: A DNA string, complement, that contains the complementary sequence of template
    constraints: both strings must be in upper case; must not contain any other characters than "ACGT"
    """
    complement = ""
    basepairdict = {"A": "T", "T": "A", "G": "C", "C": "G"}
    for c in template:
        complement += basepairdict[c]
    return complement

def dnareverse(template):
    """
    input: A DNA string, template
    output: A DNA string, reverse, that contains the reverse sequence of template
    constraints: both strings must be in upper case
    """    
    reverse = template[::-1]
    return reverse

def dnareversecomplement(template):
    """
    input: A DNA string, template
    output: A DNA string, reversecomplement, that contains the reverse complement sequence of template
    constraints: both strings must be in upper case
    """ 
    reversecomplement = dnareverse(dnacomplement(template))
    return reversecomplement

def dnareversecomplement_complete():
    """
    input: None; user is prompted for file or pasted string/values that contains a DNA sequence
    output: Reverse complement of DNA sequence
    constraints: length of DNA sequence must not exceed 1000 bp; string must not contain any other characters than "ACGT"
    """
    sequence = getinput()
    if sequence == "ERROR":
        print "No sequence to analyze.  Please retry."
    elif len(sequence) > 1000:
        print "Sequence must not exceed 1000 bp."
    else:
        newsequence = sequence.upper().strip()
        print dnareversecomplement(newsequence)
        
dnareversecomplement_complete()

# Elegant 3 Line Solution:
# st = "AAAACCCGGT"
# st = st.replace('A', 't').replace('T', 'a').replace('C', 'g').replace('G', 'c').upper()[::-1]
# print st
```

```python
# Rosalind FIB: Rabbits and Recurrence Relations

def rabbitfibonacci(n, k):
    """
    Input: 2 variables: months, n, and litter size, k
    Output: number of rabbit pairs, RP, after n months
    RP(n, k) = {n = 0: 0, n = 1: 1, n > 1: RP(n - 1, k) + RP(n - 2, k) * k}
    """
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return rabbitfibonacci(n - 1, k) + rabbitfibonacci(n - 2, k) * k
    
def rabbitfibonacci_complete():
    """
    Input: None; user is prompted for file or pasted string that contains values for n and k
    Output: number of rabbit pairs, RP, after n months
    Constraints: n <= 40, k <= 5
    """
    values = getinput().split()
    n = int(values[0])
    k = int(values[1])
    # print "n = " + n + ", k = " k
    if n > 40:
        print "n must be <= 40."
    elif k > 5:
        print "k must be <= 5."
    else:
        print rabbitfibonacci(n, k)
    
rabbitfibonacci_complete()
```

```python
# Rosalind GC: Computing GC Content

def validate_base_sequence(base_sequence, RNAflag):
    """
    Input: DNA sequence, base_sequence
    Output: True if the string base_sequence contains only upper- or lowercase T, C, A, and G characters, otherwise False
    Constraints: RNAflag checks upper and lower case U instead of T
    """
    seq = base_sequence.upper()
    return len(seq) == (seq.count('U' if RNAflag else 'T') + seq.count('G') + seq.count('A') + seq.count('C'))

def gc_content(base_seq):
    """
    Input: DNA Sequence, base_seq
    Output: the fraction of G and C characters in base_seq
    Constraint: Only one FASTA item is returned; absolute error is currently set to 0.001
    """
    # setting an absolute error of 0.001
    error = 0.001
    
    assert validate_base_sequence(base_seq, False), 'argument has invalid characters'
    seq = base_seq.upper()
    answer = float(seq.count('G') + seq.count('C'))/float(len(seq))
    
    return answer

def gc_compare(fasta_dict):
    """
    Input: A dictionary, sequence_set, that contains: key = sequencing sample ID, value = GC Content
    Output: A dictionary, output_dict: key, value = FASTA ID, %GC.  Note that there will be multiple items in output_dict if there are FASTA entries with identical %GC.
    """
    fasta_dict_gc = {}
    to_sort = []
    highest = None
    output_dict = {}
    
    # populate dictionary, fasta_dict_gc: key, value = FASTA ID, %GC
    for key, value in fasta_dict.items():
        fasta_dict_gc[key] = gc_content(value) * 100
    
    # populate list, to_sort with %GC Values and sort them
    for key, value in fasta_dict_gc.items():
        to_sort.append(value)
    
    to_sort = sorted(to_sort)
    
    # identify highest value; 
    highest = to_sort[-1]
    
    # populate output_dict
    for key, value in fasta_dict_gc.items():
        if value == highest:
            output_dict[key] = value
                
    return output_dict
    
def parse_fasta(string):
    """
    Input: A string output from getinput(), or a string of FASTA data
    Output: A dictionary, fasta_dict{fastaID: sequence}; key, value are both strings
    """ 
    fasta = []
    fasta_dict = {}
    empty_string = ""
    counter = -1

    # creates items in list, fasta: items = each FASTA sample
    for c in string:
        if c == ">":
            fasta.append(empty_string)
            counter += 1
        elif c != ">":
            fasta[counter] += c

    # create items in dictionary, fasta_dict: key, value = FASTA ID, Sequence        
    for item in fasta:
        pair = item.split("\n", 1)
        fasta_dict[pair[0]] = pair[1] 
    
    # remove EOL and whitespace
    for item in fasta_dict:
        fasta_dict[item] = fasta_dict[item].replace("\n", "")
    
    return fasta_dict

def computegccontent_complete():
    """
    Input: None; user is queried for DNA strings in FASTA format.
    Output: The ID of the string having the highest GC-content, followed by the GC-content of that string. 
    Constraint: Max 10 DNA Strings, each no greater than 1Kb.  
    """
    string = getinput()
    dictionary = parse_fasta(string)
    solution = gc_compare(dictionary)
    
    for key, value in solution.items():
        print key
        print str(value)
        
computegccontent_complete()
```

```python
# Rosalind HAMM: Counting Point Mutations

def parse_two_sequences(s):
    """
    Input: A string, s, returned from getput(), that contains two sequences of equal length
    Output: Two strings of identical length, template and mutant
    """
    strings = s.split()
    template = strings[0]
    mutant = strings[1]
    
    return (template, mutant)
    
def calculate_hamming_distance(tuple):
    """
    Input: A tuple that contains two strings of identical length, template and mutant
    Output: Integer value denoting the number of mismatched base pairs.
    Constraints: each string must be <= 1Kbp
    """
    template = tuple[0]
    mutant = tuple[1]
    assert len(mutant) == len(template), 'template and mutant must be of identical length'
    assert len(mutant) <= 1000, 'DNA sequences cannot exceed 1000 bp'
    
    counter = 0
    
    for i in range(0, len(template)):
        if template[i] != mutant[i]:
            counter += 1
            
    return counter

def calculate_hamming_distance_complete():
    """
    Input: None; User is queried for string that contains two DNA sequences
    Output: Integer value that denotes the Hamming Distance between two DNA sequences
    """
    dH = calculate_hamming_distance(parse_two_sequences(getinput()))
    
    print "dH(template, mutant) = " + str(dH)

calculate_hamming_distance_complete()

# Short 3 Line Solution:
# s1 = 'GAGCCTACTAACGGGAT'
# s2 = 'CATCGTAATGACGGCCT'
# print [ a!=b for (a, b) in zip(s1, s2)].count(True)
```

```python
# Rosalind IPRB: Mendel's First Law

"""A solution to the Rosalind prolbem "Mendel's First Law."
http://rosalind.info/problems/iprb/
"""

from __future__ import division

import sys
import unittest


class Test(unittest.TestCase):
    SAMPLE_DATASET = (2, 2, 2)
    SAMPLE_OUTPUT = 0.78333

    def test_definition_sample(self):
        self.assertAlmostEqual(
            self.SAMPLE_OUTPUT,
            probability(*self.SAMPLE_DATASET),
            places=5
        )


def probability(k, m, n):
    """Find the probability that two randomly selected mating organisms will
    produce an individual with a dominant allele."""

    pops = {'k': k, 'm': m, 'n': n}

    def _total_pop(populations):
        return sum(v for k, v in populations.items())

    def _event_probablity(subpop, populations):
        return populations[subpop] / _total_pop(populations)

    def _pick(p1, p2, populations):
        _pop = dict(populations)

        first_event_probability = _event_probablity(p1, _pop)
        _pop[p1] = _pop[p1] - 1
        second_event_probability = _event_probablity(p2, _pop)

        if 'k' in (p1, p2):
            multiplier = 1.0
        elif p1 == p2 and p1 == 'm':
            multiplier = 0.75
        elif 'm' in (p1, p2) and 'n' in (p1, p2):
            multiplier = 0.5
        elif 'n' == p1 and 'n' == p2:
            multiplier = 0.0

        return first_event_probability * second_event_probability * multiplier

    return sum(_pick(subpop1, subpop2, pops)
               for subpop2 in pops.keys()
               for subpop1 in pops.keys())


def main():
    with open(sys.argv[1]) as f:
        k, m, n = tuple(int(pop, 10)
                        for pop in f.read().replace('\n', '').split(' '))
        print probability(k, m, n)


if __name__ == '__main__':
    main()
```
